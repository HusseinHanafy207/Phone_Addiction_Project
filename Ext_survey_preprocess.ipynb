```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#read the data
df = pd.read_csv('Social_media_usage_and_mental_health.csv',encoding='utf-8')
```

```python
df.describe()
```

### Renaming Columns

```python
column_names = df.columns.tolist()
```

```python
column_names
```

```python
new_column_names = [
    'timestamp',
    'age',
    'gender',
    'relationship_status',
    'occupation_status',
    'org_affiliation',
    'use_social_media',
    'platforms_used',
    'avg_social_time',
    'purposeless_use_freq',
    'distracted_by_sm',
    'restless_without_sm',
    'easily_distracted',
    'worry_level',
    'concentration_difficulty',
    'comparison_freq',
    'comparison_feelings',
    'seek_validation',
    'feel_depressed',
    'interest_fluctuation',
    'sleep_issues'
]

df.columns = new_column_names
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['age'], vert=False)
```

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]
```

```python
plt.boxplot(df['age'], vert=False)
```

```python
df.info()
```

```python
df.drop(['timestamp','org_affiliation','platforms_used'], axis=1 ,  inplace=True)
```

```python
df.head()
```

```python
df['occupation_status'].value_counts()
```

```python
df.drop(columns=['occupation_status'], inplace=True)
```

```python
df['avg_social_time'].value_counts()
```

```python
hours_mapping = {'Less than an Hour': 1,
                'Between 1 and 2 hours': 2,
                'Between 2 and 3 hours': 3,
                'Between 3 and 4 hours': 4,
                'Between 4 and 5 hours': 5,
                'More than 5 hours': 6}
df['avg_social_time'] = df['avg_social_time'].map(hours_mapping)
```

```python
df.head()
```

```python
df.head()
```

#### Scalling

```python
'''------------------------------- Scaling Starts -----------------------------------------------'''
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

numerical_cols = ['age']

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

'''------------------------------- Scaling Ends -----------------------------------------------'''
```

```python
df['gender'].value_counts()
```

```python
df = df[df['gender'].isin(['Male', 'Female'])]
```

```python
df['gender'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['gender'], drop_first=False)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['relationship_status'], drop_first=True)
```

```python
df['use_social_media'] = df['use_social_media'].map({'Yes': 1, 'No': 0})
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.to_csv('cleaned_ext_survey_data.csv', index=False)
```
